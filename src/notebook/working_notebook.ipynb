```python
from PIL import Image
import os

def show_img(toi: int, sector: int, exp_name: str):
    toi = str(toi)
    sector = str(sector)
    img_path = f"../output/{exp_name}/images/"
    if len(toi)==3:
        toi = "0" + toi
    if len(sector)==1:
        sector = "0" + sector
    else:
        sector = str(sector)
    
    filename = f"TOI{toi}_SECTOR{sector}.png"
    imgPIL = Image.open(os.path.join(img_path, filename))
    imgPIL.show()
    
def show_tls_img(toi: int, sector: int, exp_name: str):
    toi = str(toi)
    sector = str(sector)
    img_path = f"../output/{exp_name}/tls_images/"
    if len(toi)==3:
        toi = "0" + toi
    if len(sector)==1:
        sector = "0" + sector
    else:
        sector = str(sector)
    
    filename = f"TIC{toi}_SECTOR{sector}.png"
    imgPIL = Image.open(os.path.join(img_path, filename))
    imgPIL.show()
```

```python
import deepdish as dd
# This is for biweight method
all_periods = []
all_SDE = []
for idx, row in df_results.iterrows():
    toi = str(row['TOI']).zfill(4)
    sector = str(row['SECTOR']).zfill(2)
    in_file = f"../output/TOI/tls_hdf5/TOI{toi}_SECTOR{sector}_Method_biweight.h5"
#     if not os.path.exists(in_file):
#         all_SDE.append(None)
#         pass
    h5d = dd.io.load(in_file)
    TICID = h5d["ticid"]
    print(toi, df_tois[df_tois['TIC ID']==TICID]['Period (days)'].values)
    periods = df_tois[df_tois['TIC ID']==TICID]['Period (days)'].values
    all_periods.append(periods)
    
    P_ob_err = df_tois[df_tois["TIC ID"]==TICID]["Period (days) err"].mean()
    power_list = h5d["power"]
    periods_list = h5d["periods"]
    
    TOI_SDE = []
    for period in periods:
        print("period: ", period)
        if period < min(periods_list) or period > max(periods_list):
            TOI_SDE.append(np.nan)
            continue
        mask = (periods_list > period - 2000 * P_ob_err) & (periods_list < period + 2000 * P_ob_err)
        SDE = max(power_list[mask])
        print("SDE: ", SDE)
        TOI_SDE.append(SDE)

    all_SDE.append(TOI_SDE)
    print("==============")

df_results["P_orbit"] = all_periods
df_results["SDE_biweight"] = all_SDE
```

```python
#This is for cosine method
all_periods = []
all_SDE = []
for idx, row in df_results.iterrows():
    toi = str(row['TOI']).zfill(4)
    sector = str(row['SECTOR']).zfill(2)
    in_file = f"../output/TOI/tls_hdf5/TOI{toi}_SECTOR{sector}_Method_cosine.h5"
    if not os.path.exists(in_file):
        all_SDE.append(np.nan)
        continue
    h5d = dd.io.load(in_file)
    TICID = h5d["ticid"]
    print(toi, df_tois[df_tois['TIC ID']==TICID]['Period (days)'].values)
    periods = df_tois[df_tois['TIC ID']==TICID]['Period (days)'].values
    all_periods.append(periods)
    
    P_ob_err = df_tois[df_tois["TIC ID"]==TICID]["Period (days) err"].mean()
    power_list = h5d["power"]
    periods_list = h5d["periods"]
    
    TOI_SDE = []
    for period in periods:
        print("period: ", period)
        if period < min(periods_list) or period > max(periods_list):
            TOI_SDE.append(np.nan)
            continue
        mask = (periods_list > period - 2000 * P_ob_err) & (periods_list < period + 2000 * P_ob_err)
        SDE = max(power_list[mask])
        print("SDE: ", SDE)
        TOI_SDE.append(SDE)

    all_SDE.append(TOI_SDE)
    print("==============")


df_results["SDE_cosine"] = all_SDE
```

```python
import ast

df_results = pd.read_csv("../output/TOI/method_detecton_SDE.csv", index_col=0)

def convert(x):
    try:
        return ast.literal_eval(x)
    except:
        return np.nan

df_results["SDE_cosine"] = df_results["SDE_cosine"].apply(convert)
df_results["SDE_biweight"] = df_results["SDE_biweight"].apply(convert)
df_results["SDE_GP_kernel=squared_exp"] = df_results["SDE_GP_kernel=squared_exp"].apply(convert)
```

### diff biweight and cosine

```python
idx = (df_results["cosine"]==1) & (df_results["biweight"]==1) 

df_bi_cos = df_results.loc[idx, ["SDE_biweight", "SDE_cosine"]]
df_bi_cos.head()
```

```python
df_bi_cos = df_bi_cos.dropna()
```

```python
all_diff = []
for _, row in df_bi_cos.iterrows():
    cur_diff = []
    SDE_bi = row["SDE_biweight"]
    SDE_cos = row["SDE_cosine"]

    for p_bi, p_cos in zip(SDE_bi, SDE_cos):
        cur_diff.append(np.array(p_bi - p_cos))
    all_diff.append(cur_diff)
```

```python
df_bi_cos["diff_bi_cos"] = all_diff
```

```python
df_bi_cos["diff_bi_cos"].hist()
```

* cosine method has higher SDE than biweight for the 60% of TOI samples with detected periods  
* cosine is only ~5 higher in SDE than biweight
* gp is only slightly better than cosine
