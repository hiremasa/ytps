```python
from PIL import Image
import os

def show_img(toi: int, sector: int):
    toi = str(toi)
    sector = str(sector)
    img_path = "../output/images/"
    if len(toi)==3:
        toi = "0" + toi
    if len(sector)==1:
        sector = "0" + sector
    else:
        sector = str(sector)
    
    filename = f"TOI{toi}_SECTOR{sector}.png"
    imgPIL = Image.open(os.path.join(img_path, filename))
    imgPIL.show()
```
